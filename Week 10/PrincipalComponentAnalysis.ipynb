```python
plt.style.use('ggplot')
plt.rcParams['figure.figsize']  = (12, 8)
```

```python
# Normal Distribiuted x and y vecctor with mean 0 and std 1
```

```python
x = np.random.normal(0, 1, 500)
y = np.random.normal(0,1,500)
X = np.vstack((x,y)).T

plt.scatter(X[:,0], X[:, 1])
plt.title('Generated Data')
plt.axis('equal')
```
